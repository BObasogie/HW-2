```python
# Import dependencies
import pandas as pd
import os
import csv
```

```python
# Read file into DataFrame
filename = os.path.join("..","Resources","ratings_and_sentiments.csv")
yelp_df = pd.read_csv(filename)
yelp_df.head()
```

```python
# Find all reviews with a number rating greater than or equal to 4.
yelp_df[yelp_df['num_rating'] >=4]
```

```python
# Find all the coffee shops with a overall sentiment greater than or equal to 4.

yelp_new_df = yelp_df[yelp_df['overall_sent'] >=4]

yelp_df[yelp_df['overall_sent'] >=4]
```

```python
# Return the count of the above query.
yelp_new_df["coffee_shop_name"].count()
```

```python
# Change the name of the review_text column to review
# Change the name of the bool_HIGH column to bool_high

yelp_df = yelp_df.rename(columns={"review_text": "review", "bool_HIGH": "bool_high"})
yelp_df.head()
```

```python
# Save your changes to the file as a pipe delimited file.
yelp_df.to_csv('ratings_and_sentiments.csv', sep = '|')
yelp_df
```
